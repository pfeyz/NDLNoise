```python
%matplotlib inline

import glob
import re

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


plt.rcParams['figure.figsize'] = 18, 6

# PARAMS_NOT_IMPLEMENTED = ['OPT', 'ItoC', 'QInv']


languages = {
    611: 'english',
    584: 'french',
    2253: 'german',
    3856: 'japanese'
}

colors = [[1-(n/7)] * 3 + [1] for n in range(5)]

for pathname in sorted(glob.glob('../data/output*.csv'),
                       key=lambda x: re.search(r'rate:(.*?)_', x).groups(1)):
    hyperparams = re.search(r'rate:(?P<rate>.*)_consrate:(?P<consrate>.*)\.csv', pathname).groupdict()
    df = pd.read_csv(pathname)
    parameter_names = [p
                       for p in df.columns 
                       if p[0].isupper()]

    for grammar_id in df.language.unique():
        fig, ax = plt.subplots(1,1)    
        for (noise_level, noise_amount) in enumerate(df.noise.unique()):
            rows = df[df.language.eq(grammar_id) & df.noise.eq(noise_amount)][parameter_names].values

            for row in rows:
                ax.scatter(x=[m + noise_level/5 for m in range(13)],
                           y=row,
                           marker='o', 
                           s=100,
                           edgecolors='gray',
                           facecolors=colors[noise_level])

        ax.set_xticks(np.arange(0, len(parameter_names)))
        grammar_str = format(grammar_id, '013b')
        xlabels = ['{}={}'.format(name, grammar_str[num])
                   for num, name in enumerate(parameter_names)]
        ax.set_xticklabels(xlabels, fontsize=16, );
        ax.set_title('{} {} rate={rate} cons={consrate} N={}\nNoise Levels {}'
                     .format(grammar_id, languages[grammar_id], len(rows), 
                             ', '.join(str(x) for x in df.noise.unique()),
                             **hyperparams))

        plt.scatter(np.arange(len(parameter_names)),
                    np.repeat(0.5, len(parameter_names)),
                    color='black',
                    alpha=0.5,
                    s=50,
                    marker='_')
        plt.grid(axis='x');
        plt.show();
```
